```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 创建一个 (6890, 3) 的示例数组，替换成你自己的数组
mesh = data['mesh']
# 创建一个 Matplotlib 3D图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 提取 x、y、z 坐标
x = mesh[:, 0]
y = mesh[:, 1]
z = mesh[:, 2]

# 创建散点图
ax.scatter(x, y, z)

# 设置坐标轴标签
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 显示图形
plt.show()
```
